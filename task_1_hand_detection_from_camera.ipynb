```python

# organize imports
import cv2
import numpy as np
```

```python
class HAND_DETECTION_FROM_VIDEO:
        
          
    #---------------------------------------------
    # To segment the region of hand in the image
    #---------------------------------------------
    # segment function takes the current frame and threshold number which will be used in thresholding the difference image
    def segment(frame , threshold=25):
    
        
        # we use absdiff to find the absolute difference between the background model and the current frame
        diff =  cv2.absdiff(frame , frame + 1)
    
        # using thresholding on the diff to reveal only the hand region and makes all the background black 
    
        thresholded = cv2.threshold(diff,threshold,255, cv2.THRESH_BINARY)[1]
    
        # using findContours to make contours on thresholded image (hand)
    
        contours , hierarchy  = cv2.findContours(thresholded.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    
    
        if len(contours)  == 0:
            return
        else:
        
            segmented = max(contours,key = cv2.contourArea)
            return (thresholded,segmented)
        
    def read_video_detect_hand(self):
        aWeight = 0.5
        # this function is to read video that contains hands
        video_name = 'hand_move.mp4'
        video = cv2.VideoCapture(video_name)
        # video : a variable captures the video from the camera of the lap
        # these variables are used to draw the rectangle of the ROI
        top ,right, bottom ,left = 10,350,225,590 

        num_frames = 0

        while video.isOpened():
            # frame : variable to read the frames of the video 
            (grapped , frame) = video.read()
    
            # to resize the frame
            #frame = cv2.resize(frame , width = 700)
    
            # get a copy from the frame
            clone = frame.copy()
    
            # getting the height and the width of the frame
            (height , width ) = frame.shape[:2]
    
            # taking out the region of interest from the frame
            roi = frame[top:bottom , right:left]
    
            # converting the ROI into GrayScale 
            gray = cv2.cvtColor(roi , cv2.COLOR_BGR2GRAY)
    
            # using GaussianBluring to minimize the high frequency components in the image 
            gray = cv2.GaussianBlur(gray , (7,7),0)
    
            #  we pass the current input frame into the segment function
            hand = segment(gray)
        
            if  hand is not None:
                # return the thresholded and segmented from the segment function
                (thresholded , segmented) = hand
                # we draw a contour (segmented) on the clone
                cv2.drawContours(clone,[segmented+(right , top)] , -1 , (0,0,255))
                # to show our thresholded image
                cv2.imshow("thresholded" , thresholded)
            
            cv2.rectangle(clone , (left,top) , (right , bottom),(0,255,0),2)
    
            num_frames += 1
        # using imshow we show a window on the screen calling it (Video Feed)
        # and in this window we show the frames with the hand detected that are already captured
            cv2.imshow('Video Feed',clone)
            # when we press on q button this window will close 
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        
        # this code it to release the camera 
        video.release()
        # this code is to destroy all windows
        cv2.destroyAllWindows()
 
        
        
        
        
        
    
```

```python
hand = HAND_DETECTION_FROM_VIDEO()
hand.read_video_detect_hand()
```
